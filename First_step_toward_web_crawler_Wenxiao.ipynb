## 1. Get webpage using *requests*

```python
from bs4 import BeautifulSoup
import requests
```

```python
head = "https://cnj.craigslist.org/search/somerset-nj/apa?lat=40.5010&lon=-74.5810&search_distance=114.8#search=1~gallery~0~0"
req = requests.get(head)
```

```python
webpage = req.text
print(webpage)
```

```python
soup = BeautifulSoup(webpage,'html.parser')
```

```python
print(soup.prettify())
```

```python
import pandas as pd
data = {"title": [], "link": []}

for li_tag in soup.find_all('li', class_='cl-static-search-result'):
    title = li_tag.find('div', class_='title').text.strip()
    link = li_tag.find('a')['href'].strip()
    data["title"].append(title)
    data["link"].append(link)

df = pd.DataFrame(data)
print(df)
```
